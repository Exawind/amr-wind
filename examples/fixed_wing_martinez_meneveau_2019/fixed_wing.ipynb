```python
import matplotlib.pyplot as plt
import netCDF4 as ncdf

plt.style.use('ggplot')
```

```python
# This example reads and plots the output file from fixed wing simulations using an actuator line

# Loop through all the cases
for case in ['ec4', 'ec2', 'ec1', 'ec05', 'ec025']:

    # Load the netcdf file
    nc_file = '/scratch/lmartine/amrwind_examples/fixed_wing/' + case + '/post_processing/actuator00000/F1.nc'
    data = ncdf.Dataset(nc_file)

    # Access the data
    wing = data['F1']
    rad = wing['xyz'][:, 1]
    uy = wing['veff'][-1, :, 2]
    cl = wing['cl'][-1, :]

    plt.figure(1)
    p = plt.plot(rad, -uy, 'o', alpha=0.36, label=case)
    c = p[0].get_color()

    plt.figure(2)
    p = plt.plot(rad, cl, 'o', alpha=0.36, label=case)
    c = p[0].get_color()
    
plt.figure(1)
plt.xlabel('radius [m]')
plt.ylabel('uy/U [-]')
plt.legend(loc='center left', bbox_to_anchor=(1.02, 0.5))

plt.figure(2)
plt.xlabel('radius [m]')
plt.ylabel('Cl [-]')
plt.legend(loc='center left', bbox_to_anchor=(1.02, 0.5))

plt.show()
    
```
